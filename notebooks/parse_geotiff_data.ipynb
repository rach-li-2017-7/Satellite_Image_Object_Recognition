```python
# Basic Parsing

import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Open a GeoTIFF file
geo_tiff_file = 'input/SerenitySatelliteImage_v0_20241222.tif'

with rasterio.open(geo_tiff_file) as dataset:
    # Print metadata
    print("CRS (Coordinate Reference System):", dataset.crs)
    print("Bounds:", dataset.bounds)
    print("Width, Height:", dataset.width, dataset.height)
    print("Number of Bands:", dataset.count)
    print("Driver:", dataset.driver)
    
    # Read the first band
    band1 = dataset.read(1)  # Read band 1 (Index starts from 1)
    
    # Plot the first band using imshow
    plt.figure(figsize=(10, 8))
    plt.title("GeoTIFF Band 1")
    img = plt.imshow(band1, cmap='terrain')
    plt.colorbar(img)  # Add colorbar
    plt.show()
```

```python
# Access Pixel Values and Metadata
## Extract Metadata
print("Metadata:", dataset.meta)
print("Transform:", dataset.transform)
print("Pixel Size (Resolution):", dataset.res)

## Access Pixel Values
### Print a subset of pixel values
print(band1[:5, :5])
```
